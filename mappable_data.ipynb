# Joins

* Tying all the usafacts data together

```python
deaths_monthly_for_join = deaths_monthly[["countyFIPS","1/31/20","2/29/20",
                                          "3/31/20","4/30/20","5/31/20","6/30/20","7/31/20",
                                          "8/31/20","9/30/20","10/31/20","11/30/20"]]
#deaths_monthly_for_join["10/31/20"].value_counts()
#deaths_monthly_for_join.iloc[56037]
```

```python
aggregated = confirmed_monthly.join(deaths_monthly_for_join, lsuffix = "", rsuffix = "_deaths")
aggregated = aggregated[['countyFIPS', 'County Name', 'State', 'stateFIPS', '1/31/20', '2/29/20',
       '3/31/20', '4/30/20', '5/31/20', '6/30/20', '7/31/20', '8/31/20',
       '9/30/20', '10/31/20', '11/30/20',
       '1/31/20_deaths', '2/29/20_deaths', '3/31/20_deaths', '4/30/20_deaths',
       '5/31/20_deaths', '6/30/20_deaths', '7/31/20_deaths', '8/31/20_deaths',
       '9/30/20_deaths', '10/31/20_deaths', '11/30/20_deaths']]
#aggregated = aggregated.fillna(0)
aggregated
```
