### Covid County By Race
from https://covidtracking.com/race

Lists, for each US county, the raw number of cases and deaths as well as per 100K.
Lists largest and second largest races by county.

```python
county_by_race = pd.read_csv("covid-county-by-race.csv")
county_by_race.isna().sum() #Data contains no NA's
county_by_race.sum() # 6,081,312 cases, so looking at other reporting this data seems to be thru about mid September 
```

```python
round(county_by_race.groupby("largestRace1").mean(), 2)
#round(county_by_race.groupby("largestRace2").mean(), 2)
```

```python
round(county_by_race.sort_values("deaths", ascending = False)[0:10],2)
```

```python
county_by_race.groupby("state").mean()
```
